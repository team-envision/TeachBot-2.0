```python
# Import the necessary libraries
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Use a different model that is available
model_id = "stabilityai/stable-diffusion-2-1-base"  # Changed model ID

# Load the pre-trained Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(model_id)

# Move model to GPU if available for faster inference
pipe.to("cuda" if torch.cuda.is_available() else "cpu")


```

```python
# Define a text prompt to generate an image
prompt = "A futuristic cityscape at sunset, vibrant colors, high detail"

# Generate the image from the text prompt
image = pipe(prompt).images[0]

# Show the generated image
image.show()


# Optionally, save the image
image.save("generated_image.png")
```
